**WEED GENERATION**

After fine-tuning the model using DreamBooth (with the custom token sks weed), it is possible generate high-resolution synthetic weed images using the code below. This script loads the DreamBooth fine-tuned model and generates visually coherent weed instances based on the learned concept. 

```python
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "models/model_weeds", 
    torch_dtype=torch.float16,
).to("cuda")

# Prompt used in the training
prompt = "sks weed"

images = pipe(prompt=prompt, num_inference_steps=300, guidance_scale=7.5, num_images_per_prompt=1, height=1024, width=1024).images

# Show all images
for i, img in enumerate(images):
    img.show(title=f"Weed {i+1}")

# Save generated images
for i, img in enumerate(images):
    img.save(f"generated_weed_{i+1}.png")
```
