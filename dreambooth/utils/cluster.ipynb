**Random Image Selection for Augmentation**

This script selects a random subset of images (8 per group) from a source directory, grouped by a common filename prefix (e.g., "05-15"). The selected images are copied into a separate folder for further processing or augmentation.

```python
import os
import shutil
import random
from collections import defaultdict

source_dir = "images"
output_dir = "images_to_modify"
os.makedirs(output_dir, exist_ok=True)

grouped = defaultdict(list)

for filename in os.listdir(source_dir):
    if filename.endswith(".png"):
        prefix = filename[:5]  # es. "05-15"
        grouped[prefix].append(filename)

total_selected = 0

for prefix, files in grouped.items():
    print(f"Group '{prefix}': {len(files)} founded images")

    random.shuffle(files)
    selected = files[:8]

    for fname in selected:
        src = os.path.join(source_dir, fname)
        dst = os.path.join(output_dir, fname)
        shutil.copy(src, dst)
    
    total_selected += len(selected)
```
